# HYPOTHESIS TESTING

```python
import numpy as np
from statsmodels.stats.weightstats import ztest
```

```python
# H0 = mu = 4000
# H1 = mu != 4000

sample_mean = 3050
population_mean = 1000 + 5*600
sigma_X = 25
sigma_cost = 5 * sigma_X
n = 25
print(f"Sample Mean: {sample_mean}")
print(f"Population_mean: {population_mean}")
print(f"Sigma X: {sigma_X}")
print(f"Sigma Cost: {sigma_cost}")
print(f"Sample Size: {n}")
```

```python
z = (sample_mean - population_mean) / (sigma_cost / (n ** 0.5))
z
```

```python
from scipy.stats import norm
from scipy import stats
```

```python
alpha = 0.05
z_critical = stats.norm.ppf(1 - alpha)
print(f"Critical_val: {z_critical}")
```

```python
p_value = 1 - stats.norm.cdf(z)
print(f"P-Value: {p_value}")
```

```python
if z > z_critical:
    print("Reject the null hypothesis")
    print("There is strong Evidence that weekly costs are higher than model predicts")
else:
    print("Fail to reject the null hypothesis")
    print("There is no strong Evidence that weekly costs are higher than model predicts")
```

# Estimation and Confidence Intervals

```python
import numpy as np

durability_list = [1.13, 1.55, 1.43, 0.92, 1.25, 1.36, 1.32, 0.85, 1.07, 1.48, 1.20, 1.33, 1.18, 1.22, 1.29]
durability_data = np.array(durability_list)
print(durability_data)
```

```python
sample_mean = np.mean(durability_data)
sample_std = np.std(durability_data, ddof=1)

print(f"Sample Mean: {sample_mean}")
print(f"Sample Standard Deviation: {sample_std}")
```

```python
from scipy.stats import t

degrees_of_freedom = len(durability_data) - 1
alpha = 1 - 0.99
critical_t_value = t.ppf(1 - alpha/2, degrees_of_freedom)
print(f"Critical t-value: {critical_t_value}")
```

```python
import math

sample_size = 10

margin_of_error = critical_t_value * (sample_std / math.sqrt(sample_size))

lower_bound = sample_mean - margin_of_error
upper_bound = sample_mean + margin_of_error

confidence_interval = (lower_bound, upper_bound)

print(f"Margin of Error: {margin_of_error}")
print(f"99% Confidence Interval: ({lower_bound:.3f}, {upper_bound:.3f})")
```

```python
known_population_std = 0.2
```

```python
from scipy.stats import norm

alpha = 1 - 0.99
critical_z_value = norm.ppf(1 - alpha/2)
print(f"Critical z-value: {critical_z_value}")
```

```python
import math

margin_of_error_known_std = critical_z_value * (known_population_std / math.sqrt(len(durability_data)))

lower_bound_known_std = sample_mean - margin_of_error_known_std
upper_bound_known_std = sample_mean + margin_of_error_known_std

confidence_interval_known_std = (lower_bound_known_std, upper_bound_known_std)

print(f"Margin of Error (Known Std): {margin_of_error_known_std}")
print(f"99% Confidence Interval (Known Std): ({lower_bound_known_std:.3f}, {upper_bound_known_std:.3f})")
```
