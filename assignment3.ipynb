```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv(r"C:\Users\gowra\Downloads\sales_data_with_discounts.csv")
df.head()
```

```python
df.info()
```

```python
df.dtypes
```

```python
num_colum = df.select_dtypes(include=np.number).columns.tolist()
num_colum
```

```python
mean_values = df[num_colum].mean()
print("Mean Values:\n", mean_values)
```

```python
median_values = df[num_colum].median()
print("Median Values:\n", median_values)
```

```python
mode_values = df[num_colum].mode().iloc[0]
print("Mode Values:\n", mode_values)
```

```python
std_values = df[num_colum].std()
print("Standard Deviation Values:\n", std_values)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
for col in num_colum:
    plt.figure(figsize=(10, 4))
    sns.histplot(data=df, x=col, kde=True)
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
```

```python
for col in num_colum:
    plt.figure(figsize=(10, 4))
    sns.boxplot(data=df, y=col)
    plt.title(f'Box Plot of {col}')
    plt.ylabel(col)
```

```python
cat_colum = df.select_dtypes(include='object').columns.tolist()
cat_colum
```

```python
for col in cat_colum:
    plt.figure(figsize=(10, 4))
    sns.countplot(data=df, x=col, order=df[col].value_counts().index)
    plt.title(f'Bar Plot of {col}')
    plt.xlabel(col)
    plt.ylabel('Count')
    plt.xticks(rotation=45)
    plt.tight_layout()
```
