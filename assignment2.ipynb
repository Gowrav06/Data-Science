11.	Create a small DataFrame manually with 10 rows and columns: 'Name', 'Age', 'City', and 'Salary'. Then:
- Use .info(), .describe()
- Select 'Name' and 'City' columns
- Drop 'City' column
- Fill any missing values in 'Salary' column with the mean
- Remove any duplicate rows


```python
data1 = {
    'Name': ['gowrav','gowrav', 'sai', 'kiran'],
    'Age': [24, 24, 23, 22],
    'City': ['hyd', 'hyd', 'bang', 'chennai'],
    'Salary': [50000, 50000, np.nan, 55000]
}
df1 = pd.DataFrame(data1)
df1
```

```python
df1.info()
```

```python
df1.describe()
```

```python
df1[['Name','City']]
```

```python
df1.drop('City', axis=1, inplace=True)
df1
```

```python
df2 = df1.fillna(df1['Salary'].mean())
df2
```

```python
df3 = df2.drop_duplicates()
df3
```
